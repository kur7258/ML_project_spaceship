```python
train[['CabinDeck', 'CabinNum', 'CabinSide']] = train['Cabin'].str.split('/', expand = True)
```

```python
# 거주 행성, 목적지, 냉동 수면 여부, VIP 여부, 객실 번호 별 차원 이동될 확률
# 거주 행성 - Europa: 주로 True, Earth: 주로 False, Mars: 큰 의미 없음
# 목적지 - 55 Cancri e: 주로 True, 나머지 큰 의미 없음
# 냉동 수면 - 수면 안 함: 주로 False, 수면 함: 주로 True
# VIP - 큰 의미 없는 feature 
# 객실 번호 - B: 주로 True, F: 주로 False, E: 주로 False, C: 주로 True, 나머지 큰 의미 없음
categorical_data=['HomePlanet', 'Destination', 'CryoSleep','VIP','CabinDeck']

fig=plt.figure(figsize=(12,12))

for i,name in enumerate(categorical_data):
    fig.add_subplot(3,2,i+1)
    ax = sns.countplot(data=train, x=name,hue='Transported')
    
    for p in ax.patches:
     ax.annotate("%.0f" % p.get_height(), (p.get_x() + p.get_width()/2., p.get_height() - 30), 
           ha='center', va='center', fontsize=10,  xytext=(0, 6), rotation = 0,
           textcoords='offset points')
plt.show()
```

```python
# 연령별 VIP 여부에 따른 차원 이동 확률 (큰 의미 없는 feature로 나타남)
plt.figure(figsize=(14,5))

violin =  sns.violinplot(data = train , x = "VIP" , y = "Age", kde=False, hue='Transported')

plt.show(violin)
```

```python
# Train과 test의 null값 drop으로 처리
train.dropna(inplace = True)
test.dropna(inplace = True)
```

```python
# Grid Search 사용
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from xgboost import XGBClassifier
model = XGBClassifier(learning_rate=0.1, random_state=0, n_jobs=-1)
param = {
    'min_child_weight':[1,15,30],
    'max_depth':[5,7,10],
    'n_estimators':[100,200,300]
}
```

```python
from sklearn.metrics import accuracy_score
grid = GridSearchCV(model, param_grid=param, cv=3, verbose = True, n_jobs=-1)
grid.fit(x, y)

print('xgboost best parameters : ', grid.best_params_)
print('xgboost best score : ', grid.best_score_)

grid_best = grid.best_estimator_
pred = grid_best.predict(x_validation)

print("xgboost : {}".format(accuracy_score(y_validation, pred)))
```
